### 2. 데이터 결합 - concatenate : 데이터 연결

```python
# 연습 데이터 생성
df1 = pd.DataFrame({
    'A' :['A0','A1','A2','A3'],
    'B' :['B0','B1','B2','B3'],
    'C' :['C0','C1','C2','C3'],
    'D' :['D0','D1','D2','D3'],},
    index =[0,1,2,3])

df1
```

```python
# 연습 데이터 생성
df2 = pd.DataFrame({
    'A' :['A4','A5','A6','A7'],
    'B' :['B4','B5','B6','B7'],
    'C' :['C4','C5','C6','C7'],
    'D' :['D4','D5','D6','D7'],},
    index =[4,5,6,7])

df2
```

```python
#concat을 이용한 df1, df2 데이터 결합
result = pd.concat([df1,df2])  # 세로 결합 : axis = 0 (default)
result
```

```python
#concat을 이용한 df1, df2 데이터 결합
result = pd.concat([df1,df2], axis = 1)  # 가로 결합 : axis = 1 
result
```

### 2. 데이터 결합 - merge : 데이터 병합

```python
# 연습 데이터 생성
left = pd.DataFrame({
    'key1' :['K0','K1','K2','K3'],
    'key2' :['K0','K1','K0','K1'],
    'A' :['A0','A1','A2','A3'],
    'B' :['B0','B1','B2','B3']})

left
```

```python
# 연습 데이터 생성
right = pd.DataFrame({
    'key1' :['K0','K1','K2','K3'],
    'key2' :['K0','K0','K0','K0'],
    'C' :['C0','C1','C2','C3'],
    'D' :['D0','D1','D2','D3']})

right
```

```python
#merge 결과
result = pd.merge(left,right)
result  
#=> merge 기본 옵션 : inner 
```

```python
#merge 결과
result = pd.merge(left,right, on=['key1','key2'])
result
#=> merge 기본 옵션 : inner (둘다 존재하는 것만)
```

```python
#merge 결과 
result = pd.merge(left,right, on=['key1','key2'],how ='left')
result
#=> merge 옵션 : left
```

```python
#merge 결과
result = pd.merge(left,right, on=['key1','key2'],how ='right')
result
#=> merge 옵션 : right
```

```python
#merge 결과
result = pd.merge(left,right, on=['key1','key2'],how ='outer')
result
#=> merge 옵션 : outer
```

```python
#merge 결과
result = pd.merge(left,right, on=['key1','key2'],how ='outer', indicator=True)
result
#=> merge 옵션 : outer, 
```

### 2. 데이터 결합 - join

```python
# 연습 데이터 생성
left = pd.DataFrame({
    'key1' :[1,2,3,4],
    'A' :['A0','A1','A2','A3'],
    'B' :['B0','B1','B2','B3']})

left
```

```python
# 연습 데이터 생성
right = pd.DataFrame({
    'C' :['C1','C2','C5'],
    'D' :['D1','D2','D5']},
    index = [1,2,5])

right
```

```python
#join 결과 
result = left.join(right)
result
# join 의 기본 옵션 : left 
```

```python
#join 결과 
result = left.join(right, how='right')
result
# join 의 옵션 : right
```

```python
#join 결과 
result = left.join(right, how ='inner')
result
# join 의 옵션 : inner 
```

```python
#join 결과 
result = left.join(right, how ='outer')
result
# join 의 옵션 : outer
```

```python
##on : join 의 key 지정 (left의 column명)

result = left.join(right, on = 'key1') # left의 인덱스열이 아닌 key1열을 기준으로 right의 열을 맞추어 join 
result
## 밑과 비교
```

```python
##위와 비교

result = left.join(right)
result
```

```python
## 참고
#merge로 결합
result2 = pd.merge(left,right,left_on='key1',right_index=True,how='left') ;
result2

#left.join(right, on = 'key1')  과 동일 
```
