# 데이터 정제
데이터 확인을 통하여 발견된 결측치나 이상치에 대한 적절한 처리를 통해 데이터 품질을 보강하는 작업

1. 분석에 필요한 데이터 구성하기

## step0 
### 패키지 불러오기

```python
#데이터 구성 패키지 : Series, DataFrame
import pandas as pd
#행렬 연산 패키지
import numpy as np
# Scaling 클래스 
from sklearn.preprocessing import scale, minmax_scale, robust_scale 

# matplotlib 이용한 Background 그래프 실행 및 결과 확인
%matplotlib inline
```

```python
# engine : 파일명 또는 데이터에 한글 포함 -> "python" 사용
df = pd.read_csv('../data/FITNESS_결측.csv', engine="python")
df.head(3)
```

## step2 
### 결측치 확인

```python
# 결측치 확인 : pandas.DataFrame.isnull() 또는 pandas.DataFrame.isna()
df.isnull().head(3)
# df.isna().head(3)
```

### 결측치 처리 방법
* 연속형 변수의 경우   
1. 평균값(mean)으로 대체  
2. 중앙값(median)으로 대체  
3. 최소, 최대값의 중앙값(mid-range)으로 대체   
4. 분포 기반 대체(자료의 랜덤화 비율 고려)  
5. 모델(회귀분석, Tree 등)에 의한 대체 (발생 유형이 유사한 값 참조)  

* 범주형 변수의 경우  
1. 최빈값(mode, 빈도가 가장많은)으로 대체  
2. 분포 기반 대체(자료의 랜덤화 비율고려)  
3. 모델(Tree등)에 의한 대체 (발생 유형이 유사한 값 참조)
  
    

*발생 원인 및 변수 특성을 파악하여 적절하게 대체하는 것이 바람직하나 무조건 대체하는 것은 분석결과를 왜곡시킬 수 있기에 제외하는 것도 고려

### 결측치 확인 - 변수별 결측 현황 확인 

```python
#결측치(True) 현황 요약 : sum 함수
df.isnull().sum()
```

### 결측치가 있는 모든 행 추출 

```python
null_data = df[df.isnull().any(axis=1)]
null_data
```

### 결측치 확인 - 범주형(문자) 변수 결측치 처리 : GENDER 변수의 결측치에 "여성" 으로 대체

```python
#문자형 결측치 처리 : pandas.DataFrame.fillna(값, inplace : 결측치 입력 후 저장 여부)
df['GENDER'].fillna("여성",inplace = True)
df.head(3)
```

=> GENDER 변수의 결측치를 "여성"으로 대체한 기준은? 이름이 여성이름(Gracie)이기에 

### 결측치 확인 - 연속형(숫자)변수 결측치 처리 : GENDER 별 변수 WEIGHT의 평균 계산 및 대체

```python
#숫자형 결측치 처리 : groupby(범주형 변수).agg("집계함수")
df.groupby("GENDER")["WEIGHT"].agg("mean")
```

### 결측치 확인 - 연속형(숫자)변수 결측치 처리 : WEIGHT 변수의 결측치를 GENDER="남성"의 WEIGHT 평균으로 대체 

```python
#숫자형 결측치 처리 : transform("집계함수")
df['WEIGHT'] = df['WEIGHT'].fillna(df.groupby("GENDER")["WEIGHT"].transform("mean")).round(3)
df.head(3)
```

```python
df.groupby("GENDER")["WEIGHT"].transform("mean")
```

## step4

### Scale 변환: Scale 변환 대상 변수 = 숫자형 변수(문자형 변수는 적용 안됨)

```python
#select_dtypes 조건으로 변수 유형 선택 : select_dtypes = "object"  문자형 변수
#숫자형 변수 선택 : select_dtypes="object"제외
df_num = df.select_dtypes(exclude ="object")
df_num.head(3)
```

```python
#문자형 변수 선택 : select_dtypes = "object" 선택
df_char = df.select_dtypes(include="object")
df_char.head(3)
```

### Sclae변환1. (평균,표준편차) : 평균 = 0, 표준편차 = 1 기준변환

```python
#Scale변환 : 평균, 표준편차 기준
df_scale_std = scale(df_num)
df_scale_std
```

```python
#Scale 적용/ 변환하면 numpy의 ndarray 형태로 저장 -> DataFrame 으로 변환
df_scale_std = pd.DataFrame(df_scale_std, columns=df_num.columns)
df_scale_std.head(3)
```

=> 평균 0, 표준편차 1 인 데이터로 변환

```python
#Scale 변환 결과 확인 : 요약 통계량 : pandas.DataFrame.describe()
df_scale_desc = df_scale_std.describe()
df_scale_desc.round(3)
```

=> 변환 데이터 요약 통계량

### Scale 변환2. (최소 - 최대) : 최소 = 0, 최대 = 1 변환

```python
#Scale 변환 : 최소, 최대값 기준
df_scale_minmax = minmax_scale(df_num)
df_scale_minmax
```

```python
df_scale_minmax = pd.DataFrame(df_scale_minmax, columns=df_num.columns)
df_scale_minmax.head(3)
```

=> 최소값 0, 최대값 1인 데이터로 변환 

```python
#Scale 변환 결과 확인 : 요약통계량 : pandas.DataFrame.describe()
df_scale_desc = df_scale_minmax.describe()
df_scale_desc.round(3)
```

=> 변환데이터 요약 통계량

### Scale 변환3. (로버스트) : 중앙값 = 0 변환(이상치 영향 감소)

```python
#Scale 변환 : 로버스트 기준
df_scale_robust = robust_scale(df_num)
df_scale_robust 
```

```python
#Scale 적용/ 변환하면 numpy의 ndarray 형태로 저장 -> DataFrame 으로 변환
df_scale_robust = pd.DataFrame(df_scale_robust, columns=df_num.columns)
df_scale_robust.head(3)
```

```python
#Scale 변환 결과 확인 : 요약통계량 : pandas.DataFrame.describe()
df_scale_desc = df_scale_robust.describe()
df_scale_desc.round(3)
```

=> 50%의 값은 모두 0인 것을 볼수있다

### Scale 변환 결과 비교 : 예시로 RSTPULSE 변수 비교 

```python
df_rstpulse = pd.DataFrame()
df_rstpulse["Raw"] = df_num["RSTPULSE"]
df_rstpulse["Standard"] = df_scale_std["RSTPULSE"]
df_rstpulse["MinMax"] = df_scale_minmax["RSTPULSE"]
df_rstpulse["Robust"] = df_scale_robust["RSTPULSE"]
df_rstpulse.head().round(3)
```
